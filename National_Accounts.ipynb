```python
from urllib.request import urlretrieve
national_accounts_url = 'https://raw.githubusercontent.com/rohanbjkr/dataset_sudal/all-data/national_accounts.csv'
urlretrieve(national_accounts_url,'national_accounts.csv')
```

```python
import pandas as pd
national_accounts_df = pd.read_csv('national_accounts.csv')
national_accounts_df
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=national_accounts_df.year,y=national_accounts_df.rgdp_producer_growth, 
                         mode = 'lines+markers',name="Purchaser's Price"))
fig.add_trace(go.Scatter(x=national_accounts_df.year,y=national_accounts_df.rgdp_basic_growth, 
                         mode = 'lines+markers',name="Basic Price"))
fig.update_layout(template='simple_white',
                  yaxis_title = "Percentage",
                  xaxis_title = "Year",
                  title = "Real GDP growth at basic and purchaser's price",
                  showlegend = False
                 )
fig.update_yaxes(tickformat=".2f")

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=national_accounts_df.year,y=national_accounts_df.agri_share, 
                         mode = 'lines+markers',name="Agriculture"))
fig.add_trace(go.Scatter(x=national_accounts_df.year,y=national_accounts_df.industry_share, 
                         mode = 'lines+markers',name="Industry"))
fig.add_trace(go.Scatter(x=national_accounts_df.year,y=national_accounts_df.services_share, 
                         mode = 'lines+markers',name="Service"))
fig.update_layout(template='simple_white',
                  yaxis_title = "Percentage",
                  xaxis_title = "Year",
                  title = "Share of Agriculture, Industry, and Service Sector in GDP of Nepal",
                  showlegend=False
                 )
fig.update_yaxes(tickformat=".2f")
# fig.update(layout_showlegend=False)

fig.show()
```
